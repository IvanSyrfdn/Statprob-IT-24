***Modus (Mode)***

```python
# Import library Counter untuk menghitung modus
from collections import Counter

#Membuat fungsi agar mudah menghitung modus tiap data
def hitungModus(data):
  dataUrut = sorted(data) # biar enak bacanya, diurutin dulu
  counter = Counter(dataUrut)
  modus_data = dict(counter) # ubah ke tipe data dictionary
  max_counter = max(list(counter.values())) # mencari data yang memiliki counter paling banyak
  print(f"Jumlah kemunculan tertinggi: {max_counter}")
  nilai_modus = []
  for k, v in modus_data.items():
    print(f"{k} muncul {v} kali")
    if v == max(list(counter.values())):
      nilai_modus.append(k)
  print(f"Modus: {nilai_modus}")
  return nilai_modus

# Intensitas Hujan
print('Intensitas Hujan')
ModusIntensitasHujan = hitungModus(data['Intensitas_Hujan'])
# Penjualan A
print('Penjualan A')
ModusPenjualanA = hitungModus(data['Penjualan_A'])
# Penjualan B
print('Penjualan B')
ModusPenjualanB = hitungModus(data['Penjualan_B'])

#simpan dulu wak
Modus = [ModusIntensitasHujan, ModusPenjualanA, ModusPenjualanB] # ga tau nyoba nyoba ternyata bisa di store di array
```

***Median***

```python
#Membuat fungsi agar mudah menghitung median tiap data
def hitungMedian(data):
  dataUrut = sorted(data)
  n = len(dataUrut)
  nilaiTengah = n//2
  if n % 2 == 0 :
    median1 = dataUrut[nilaiTengah-1]
    median2 = dataUrut[nilaiTengah]
    median = (median1+median2)/2
  else :
    median = dataUrut[nilaiTengah]
  print(f'Median: {median}')
  return median

# Intensitas Hujan
print('Intensitas Hujan')
medianIntensitasHujan = hitungMedian(data['Intensitas_Hujan'])
# Penjualan A
print('Penjualan A')
medianPenjualanA = hitungMedian(data['Penjualan_A'])
# Penjualan B
print('Penjualan B')
medianPenjualanB = hitungMedian(data['Penjualan_B'])

#simpan dulu wak
median = [medianIntensitasHujan, medianPenjualanA, medianPenjualanB]
```

***Interquartile Range (IQR)***

```python
#import library
import numpy as np
def hitungIQR(data):
  q1 = np.percentile(data, 25)
  q3 = np.percentile(data, 75)
  iqr = q3 - q1
  print("Interquartile Range (IQR):", iqr)
  return q1, q3, iqr

# Intensitas Hujan
print('Intensitas Hujan')
iqrIntensitasHujan = hitungIQR(data['Intensitas_Hujan'])
# Penjualan A
print('Penjualan A')
iqrPenjualanA = hitungIQR(data['Penjualan_A'])
# Penjualan B
print('Penjualan B')
iqrPenjualanB = hitungIQR(data['Penjualan_B'])

#simpen datanya wak
iqrPaling = [iqrIntensitasHujan, iqrPenjualanA, iqrPenjualanB]
q1A, q3A, iqr1 = iqrPaling[0]
q1B, q3B, iqr2 = iqrPaling[1]
q1C, q3C, iqr3 = iqrPaling[2]
```

***Variansi***

```python
#Membuat fungsi agar mudah menghitung range tiap data
def hitungMean(data):
  jumlah = round(sum(data), 2)
  n = len(data)
  mean = round(jumlah/n, 2)
  return mean, n

def hitungVariansi(data):
  mean, n = hitungMean(data)
  jumlahSelisihKuadrat = 0
  for x in data :
    selisihKuadrat = (x - mean)**2
    jumlahSelisihKuadrat += selisihKuadrat
  variansi = round(jumlahSelisihKuadrat/n, 2)
  print(f'Variansi: {variansi}')
  return variansi

# Intensitas Hujan
print('Intensitas Hujan')
hitungVariansi(data['Intensitas_Hujan'])
# Penjualan A
print('Penjualan A')
hitungVariansi(data['Penjualan_A'])
# Penjualan B
print('Penjualan B')
hitungVariansi(data['Penjualan_B'])
```

***Standar Deviasi***

```python
#Membuat fungsi agar mudah menghitung range tiap data
def hitungMean(data):
  jumlah = round(sum(data), 2)
  n = len(data)
  mean = round(jumlah/n, 2)
  return mean, n

def hitungVariansi(data):
  mean, n = hitungMean(data)
  jumlahSelisihKuadrat = 0
  for x in data :
    selisihKuadrat = (x - mean)**2
    jumlahSelisihKuadrat += selisihKuadrat
  variansi = round(jumlahSelisihKuadrat/n, 2)
  print(f'Variansi: {variansi}')
  return variansi

def hitungStdDeviasi(data):
  variansi = hitungVariansi(data)
  stdDeviasi = round(variansi**0.5, 2) #sqrt = pangkat 1/2
  print(f'Standar Deviasi: {stdDeviasi}')

# Intensitas Hujan
print('Intensitas Hujan')
hitungStdDeviasi(data['Intensitas_Hujan'])
# Penjualan A
print('Penjualan A')
hitungStdDeviasi(data['Penjualan_A'])
# Penjualan B
print('Penjualan B')
hitungStdDeviasi(data['Penjualan_B'])
```

***Skewness***

```python
#Membuat fungsi agar mudah menghitung range tiap data
def hitungMean(data):
  jumlah = round(sum(data), 2)
  n = len(data)
  mean = round(jumlah/n, 2)
  return mean, n

def hitungVariansi(data):
  mean, n = hitungMean(data)
  jumlahSelisihKuadrat = 0
  for x in data :
    selisihKuadrat = (x - mean)**2
    jumlahSelisihKuadrat += selisihKuadrat
  variansi = round(jumlahSelisihKuadrat/n, 2)
  print(f'Variansi: {variansi}')
  return variansi

def hitungStdDeviasi(data):
  variansi = hitungVariansi(data)
  stdDeviasi = round(variansi**0.5, 2)
  print(f'Standar Deviasi: {stdDeviasi}')
  return stdDeviasi

def hitungSkewness(data):
  mean, n = hitungMean(data)
  stdDeviasi = hitungStdDeviasi(data)
  jumlah = 0
  for x in data :
    selisih = ((x - mean)/stdDeviasi)**3
    jumlah += selisih
  skew = round(jumlah/n, 2)
  print(f'Skewness: {skew}')

# Intensitas Hujan
print('Intensitas Hujan')
hitungSkewness(data['Intensitas_Hujan'])
# Penjualan A
print('Penjualan A')
hitungSkewness(data['Penjualan_A'])
# Penjualan B
print('Penjualan B')
hitungSkewness(data['Penjualan_B'])
```

***Kurtosis***

```python
#Membuat fungsi agar mudah menghitung range tiap data
def hitungMean(data):
  jumlah = round(sum(data), 2)
  n = len(data)
  mean = round(jumlah/n, 2)
  return mean, n

def hitungVariansi(data):
  mean, n = hitungMean(data)
  jumlahSelisihKuadrat = 0
  for x in data :
    selisihKuadrat = (x - mean)**2
    jumlahSelisihKuadrat += selisihKuadrat
  variansi = round(jumlahSelisihKuadrat/n, 2)
  print(f'Variansi: {variansi}')
  return variansi

def hitungStdDeviasi(data):
  variansi = hitungVariansi(data)
  stdDeviasi = round(variansi**0.5, 2)
  print(f'Standar Deviasi: {stdDeviasi}')
  return stdDeviasi

def hitungKurtosis(data):
  mean, n = hitungMean(data)
  stdDeviasi = hitungStdDeviasi(data)
  jumlah = 0
  for x in data :
    selisih = ((x - mean)/stdDeviasi)**4
    jumlah += selisih
  kurt = round(jumlah/n, 2)
  print(f'Kurtosis: {kurt}')

# Intensitas Hujan
print('Intensitas Hujan')
hitungKurtosis(data['Intensitas_Hujan'])
# Penjualan A
print('Penjualan A')
hitungKurtosis(data['Penjualan_A'])
# Penjualan B
print('Penjualan B')
hitungKurtosis(data['Penjualan_B'])
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import math
```

```python
data = pd.read_excel('/Dataset.xlsx', header=0)
data
"""
Column name mapping

Tanggal          ->  date
Curah Hujan (mm) ->  rainfall
Penjualan A      ->  sale_a
Penjualan B      ->  sale_b
"""
data.columns = ['date', 'rainfall', 'sale_a', 'sale_b']
data
```

```python
def get_frequency_distribution(data):
    if type(data) == pd.Series:
        data = data.array

    n = len(data)
    mx, mi = data.max(), data.min()
    R = mx - mi
    k = math.ceil(1 + 3.3 * math.log(n, 10))
    I = math.ceil(R / k)

    classes = []
    ranges  = []
    edge_classes = []
    for i in range(1, k + 1):
        upper = mi + (i - 1) * I
        lower = upper + I - 1
        edge = upper - 0.5
        classes.append(i)
        edge_classes.append(edge)
        ranges.append(f'{upper} - {lower}')
    edge_classes.append(lower + 0.5)

    df = pd.DataFrame(data, columns=['value'])
    df["range"] = pd.cut(df["value"], bins=edge_classes, labels=ranges, include_lowest=True)
    frequency_table = df["range"].value_counts().sort_index()
    relative_frequency = (frequency_table / frequency_table.sum() * 100).round(2)
    midpoints = [
        (edge_classes[i] + edge_classes[i + 1]) / 2 for i in range(len(edge_classes) - 1)
    ]
    frequency_distribution = pd.DataFrame(
        {
            "class": classes,
            "range": frequency_table.index,
            "frequency": frequency_table.values,
            "relative": relative_frequency,
            "midpoint": midpoints,
        }
    )
    frequency_distribution.reset_index(drop=True, inplace=True)

    return frequency_distribution
```

```python
print(" FREQUENCY DISTRIBUTION TABLE PRODUK A ".center(50, "-"))
freq_a = get_frequency_distribution(data['sale_a'])
print(freq_a.to_string(index=False))
print(" FREQUENCY DISTRIBUTION PRODUK A".center(50, "-"))
```

```python
print(" FREQUENCY DISTRIBUTION TABLE PRODUK B ".center(50, "-"))
freq_b = get_frequency_distribution(data['sale_b'])
print(freq_b.to_string(index=False))
print(" FREQUENCY DISTRIBUTION PRODUK B".center(50, "-"))
```

***OUTLIERS dan KESIMPULAN***

Outlier adalah nilai yang berbeda secara signifikan dari nilai lain dalam kumpulan data.

Pada Dataset ini terdapat *outlier* di masing-masing data penjualan A dan penjualan B. *outlier* ini berupa penjualan yang berjumlah 0 pada A dan B. Jumlah ini sangat berbeda secara signfikan dibandingkan dengan data yang lainnya.

Kesimpulan

Berdasarkan Dataset yang ada bisa ditarik kesimpulan bahwa
Data A memiliki:

*   mean= 131.83
*   median = 52
*   nilai kuartil 1 = 5.0
*   kuartil 3 = 238.0
*   Standar Deviasi A = 141.01
*   Modus B = 0

Data B memiliki:

*   mean= 205.55
*   median = 157
*   nilai kuartil 1 = 80.0
*   nilai kuartil 3 = 324.0
*   Standar Deviasi A = 158.9
*   Modus B = 0

Masing-masing data A dan B sama-sama miring ke kanan.
